# Exploratory Data Analysis (EDA)

### Project 1: General EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Global_Superstore.xlsx")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Sales'].fillna(df['Sales'].median(), inplace=True)
df['Profit'].fillna(df['Profit'].median(), inplace=True)
df['Region'].fillna('Unknown', inplace=True)
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['Sales'] >= lower) & (df['Sales'] <= upper)]
```

```python
df[['Sales', 'Profit']].describe()
```

```python
correlation = df[['Sales', 'Profit', 'Quantity', 'Discount']].corr()
correlation
```

```python
plt.figure()
plt.hist(df['Sales'], bins=30)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Profit'])
plt.title("Profit Boxplot")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure()
sns.barplot(x='Region', y='Sales', data=df)
plt.xticks(rotation=45)
plt.title("Sales by Region")
plt.show()
```
