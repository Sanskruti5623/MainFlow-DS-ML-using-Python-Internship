# Task 1: Data Analysis Project Using Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student-mat.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.drop_duplicates()
```

```python
avg_g3=df["G3"].mean()
avg_g3
```

```python
stu_abv_15=df[df["G3"]>15].shape[0]
stu_abv_15
```

```python
correlation=df["studytime"].corr(df["G3"])
correlation
```

```python
gender_avg=df.groupby("sex")["G3"].mean()
gender_avg
```

```python
plt.hist(df["G3"])
```

```python
plt.hist(df["G3"],bins=6)
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.title("Final Grade")
```

```python
plt.scatter(df['studytime'], df['G3'], alpha=0.7)
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')
plt.title('Study Time vs Final Grade')
plt.show()
```

```python
avg_scores = df.groupby('sex')['G3'].mean()

plt.figure(figsize=(6, 5))
avg_scores.plot(kind='bar', color=['steelblue', 'salmon'])
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.title('Average Final Grade by Gender')
plt.xticks(rotation=0)
plt.show()
```
